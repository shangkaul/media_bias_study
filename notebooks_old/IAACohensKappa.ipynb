```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
```

```python
df = pd.read_csv('annotated_articles.csv')
```

```python
column_mapping = {
    '1= Yes, 0 = No, 2 = Not Sure' : 'Article',
    'Unnamed: 1' : 'Newssite',
    'Unnamed: 2' : 'URL',
    'Unnamed: 3' : 'Oishani Relevant?',
    'Unnamed: 4' : 'Matt Relevant?'
}

df.rename(columns=column_mapping, inplace=True)
```

```python
df.drop(index=0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df[df['Oishani Relevant?'] != df['Matt Relevant?']]
```

```python
df['Newssite'].nunique()
```

```python
df.dtypes
df['Oishani Relevant?'] = pd.to_numeric(df['Oishani Relevant?'])
df['Matt Relevant?'] = pd.to_numeric(df['Matt Relevant?'])
```

```python
cohen_kappa_score(df['Matt Relevant?'], df['Oishani Relevant?'])
```

```python
df.to_csv('annotated_articles.csv')
```
